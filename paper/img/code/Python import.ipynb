```python
%pylab inline

plt.rcParams['figure.figsize'] = (12, 16)
plt.rcParams['font.size'] = 14.0

import scipy.io as sio
from math_biol import *
```

```python
saveFigs = False
filenames = ['stable.mat','cyclic.mat', 'chaos.mat']
arrangement = ['311', '312', '313']
titles = ['Stable attractor', 'Cyclic attractor', 'Chaotic attractor']
labels = ['A', 'B', 'C']
```

```python
counter = 0
for filename in filenames:
    (y_out, t_out) = import_timeseries(filename)
    plt.subplot(arrangement[counter])
    #plt.subplots_adjust(hspace=0.6)
    plt.plot(t_out, y_out)
    plt.title(titles[counter])
    plt.ylabel('Population')
    
    ax = gca()
    ax.text(1.05, 0.50, labels[counter], transform=ax.transAxes, fontsize=16, fontweight='bold', va='top', ha='right')
    
    counter += 1
    
plt.xlabel('Time');

if saveFigs:
    plt.savefig('../ts.png')
```
